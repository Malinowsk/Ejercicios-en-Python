```python
import json  #Importar las funciones de json en la 
#clase 15 entenderemos mejor el IMPORT

data = {}

data['clients'] = []

data['clients'].append({
    'first_name': 'Sigrid',
    'last_name': 'Mannock',
    'age': 27,
    'amount': 7.17})

data['clients'].append({
    'first_name': 'Joe',
    'last_name': 'Hinners',
    'age': 31,
    'amount': [1.90, 5.50]})

data['clients'].append({
    'first_name': 'Theodoric',
    'last_name': 'Rivers',
    'age': 36,
    'amount': 1.11})

```

```python
import pprint  # significa pretty print
pprint.pprint(data)
```

```python
# libreria JSON sirve para transformar diccionarios en formato JSON o viceversa
# tiene dos comandos
# comando dump que sirve para volcar un diccionario en un arcivo
# comando load que sirve para leer un archivo JSON y retornarlo como un diccionario de Python
```

```python
# Guardar el diccionario en un archivo JSON
# el comando with, cierra el archivo por nosotros, no tenemos hacer .close()
with open(ruta + "/primerJson.json", 'w') as file:
    json.dump(data, file, indent=4)
```

```python
# Hacemos esto para leer el JSON y convertirlo en un diccionario
with open(ruta + "/primerJson.json", "r") as file:
    dataLectura = json.load(file)
```

```python
dataLectura
```

```python
for client in dataLectura['clients']:
    print('First name:', client['first_name'])
    print('Last name:', client['last_name'])
    print('Age:', client['age'])
    print('Amount:', client['amount'])
    print('')
```
