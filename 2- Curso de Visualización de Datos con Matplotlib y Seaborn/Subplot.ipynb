```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)       # 1 fila, dos columnas y se esta referenciando el número 1
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')       # Con la referencia aún al plot 1, se dibuja otra gŕafica
plt.subplot(1,2,2)       # ahora se referencia el plot número 2
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
